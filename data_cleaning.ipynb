#Initial data exploration revealed missing values in key columns such as #director (~30%), cast, and country. The dataset also required type conversion for date fields and text preprocessing for categorical variables.

```python
df.isnull().sum()
```

<h1> Cleaning </h1>

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['rating'] = df['rating'].fillna('Not Rated')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added'])
```

```python
df.sample(5)
```

```python
df['title'] = df['title'].str.strip()
df['country'] = df['country'].str.strip()
df['director'] = df['director'].str.strip()
```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['type'].value_counts()
df['country'].value_counts().head(10)
df['rating'].value_counts()
```
